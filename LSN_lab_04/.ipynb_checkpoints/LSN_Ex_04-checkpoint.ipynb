# Laboratorio di simulazioni numeriche
### Esercizio 4.1
All'interno del codice di dinamica molecolare fornito, viene richiesto di implementare un metodo alternativo (metodo 1) per l'inizializzazione delle posizioni delle particelle che soddisfi i seguenti requisiti:
- sia selezionabile fornendo come input una variabile booleana;
- legga in ingresso posizioni $\vec{r}(t-\delta t)$ e $\vec{r}(t)$ e le utilizzi per calcolare $\vec{r}(t+\delta t)$;
- calcoli le velocità $\vec{v}(t+\delta t/2)$ e le riscali in funzione della temperatura data in input $T^*$;
- utilizzi le velocità riscalate per ottenere nuove posizioni $\vec{r}_{new}(t)$, utilizzabili insieme a $\vec{r}(t+\delta t)$ per avviare la simulazione.

Tale metodo di inizializzazione deve essere confrontato con quello fornito (metodo 0), il quale prevede in ingresso unicamente $\vec{r}(t)$ e calcola $\vec{r}(t+\delta t)$ utilizzando velocità random.

Di seguito vengono riportati grafici per la tempertaura, energia cinetica, potenziale e totale del sistema di particelle avendo utilizzato i due metodi. Entrambe le simulazioni sono state eseguite a partire da posizioni $\vec{r}(t)$ coerenti con la fase liquida (fase a cui si riferiscono i parametri forniti in ingresso): in entrambi i casi l'equilibrio termodinamico è raggiunto già dai primi passi della simulazione. Tuttavia se nel metodo 0 vengono fornite posizioni poco probabili per la fase liquida (ad esempio un cristallo FCC) il processo di raggiungimento dell'equilibrio non è immediato e richiede alcuni restart tramite metodo 1.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1000)

T_0 = np.loadtxt("Lab4/output_temp_r.dat")
K_0 = np.loadtxt("Lab4/output_ekin_r.dat")
U_0 = np.loadtxt("Lab4/output_epot_r.dat")
E_0 = np.loadtxt("Lab4/output_etot_r.dat")

T_1 = np.loadtxt("Lab4/output_temp_2.dat")
K_1 = np.loadtxt("Lab4/output_ekin_2.dat")
U_1 = np.loadtxt("Lab4/output_epot_2.dat")
E_1 = np.loadtxt("Lab4/output_etot_2.dat")


fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(16, 6))
method_0, method_1 = axes.flatten()

method_0.plot(n,T_0,label="Temperatura")
method_0.plot(n,K_0,label="Energia cinetica")
method_0.plot(n,U_0,label="Energia potenziale")
method_0.plot(n,E_0,label="Energia totale")
method_0.set_title('Metodo 0')
method_0.set(xlabel='#print')
method_0.grid(True)
method_0.legend(bbox_to_anchor=(0.65, 0.5))

method_1.plot(n,T_1,label="Temperatura")
method_1.plot(n,K_1,label="Energia cinetica")
method_1.plot(n,U_1,label="Energia potenziale")
method_1.plot(n,E_1,label="Energia totale")
method_1.set_title('Metodo 1')
method_1.set(xlabel='#print')
method_1.grid(True)
method_1.legend(bbox_to_anchor=(0.65, 0.5))

plt.show()
```

Il comportamento qualitativo delle simulazioni svolte con i due metodi è lo stesso. Si osserva che l'energia totale del sistema è approssimativamente costante in entrambi i casi. Ciò è conseguenza del fatto che il codice si riferisce a un ensemble microcanonico.

### Esercizio 4.3
Nel caso specifico dell'argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu), vanno studiati i valori medi delle quantità fisiciche dell'Es. 4.2 (espressi in SI) nelle condizioni di:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$, $r_c = 2.2$;
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$, $r_c = 2.5$;
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$, $r_c = 5.0$;

dove $\rho^\star$ è la desità delle particelle e $r_c$ è il raggio di cut-off dell'interazione.

I risultati delle simulazioni riportati di seguito si riferiscono a posizioni iniziali ottenute attraverso simulazioni svolte in precedenza. Tali simulazioni sono state svolte coerentemente con la fase studiata, ma con posizioni in ingresso proprie di un cristallo FCC e con un numero di passi 5 volte maggiore di quello qui riportato. Si è voluto operare in questo modo per evitare effetti legati al non raggiungimento della condizione di equilibrio. Le simulazioni di partenza sono stata svolte con il metodo 0 per l'inizializzazione delle velocità, il restart con il metodo 1.

Il processo di equilibrazione non è riportato di seguito. Osserviamo tuttavia che eseguendo l'equilibrazione con le temperature di ingresso fornite, il sistema si stabilizza a temperature che possono variare rispetto a quelle attese. In effetti lavorando con un ensemble microcanonico, anche all'equilibrio si osservano oscillazioni della temperatura più o meno vistose a seconda della fase in esame. Un possibile trucco per raggiungere la temperatura target può essere quello di utilizzare una temperatura più bassa nelle prime fasi dell'equilibrazione.


### Fase solida

```python
T = np.loadtxt("Lab4/ave_temp_s.out")
Ekin = np.loadtxt("Lab4/ave_ekin_s.out")
Epot = np.loadtxt("Lab4/ave_epot_s.out")
Etot = np.loadtxt("Lab4/ave_etot_s.out")

nblock = range(0, 100, 1)
T_m = T[:,0] *120
T_err = T[:,1] *120
Ekin_m = Ekin[:,0] *120 *1.38 * 10**(-23)
Ekin_err = Ekin[:,1] *120 *1.38 * 10**(-23)
Epot_m = Epot[:,0] *120 *1.38 * 10**(-23)
Epot_err = Epot[:,1] *120 *1.38 * 10**(-23)
Etot_m = Etot[:,0] *120 *1.38 * 10**(-23)
Etot_err = Etot[:,1] *120 *1.38 * 10**(-23)

x, y = [0,100], [96,96]

plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.errorbar(nblock,T_m,yerr=T_err)
plt.plot(x,y,color='r')
plt.xlabel('#block')
plt.ylabel('<T> [K]')
plt.title('Temperatura media')
plt.grid(True)

plt.subplot(2,2,2)
plt.errorbar(nblock,Ekin_m,yerr=Ekin_err)
plt.xlabel('#block')
plt.ylabel('<Ekin> [J]')
plt.title('Energia cinetica media')
plt.grid(True)

plt.subplot(2,2,3)
plt.errorbar(nblock,Epot_m,yerr=Epot_err)
plt.xlabel('#block')
plt.ylabel('<Epot> [J]')
plt.title('Energia potenziale media')
plt.grid(True)

plt.subplot(2,2,4)
plt.errorbar(nblock,Etot_m,yerr=Etot_err)
plt.xlabel('#block')
plt.ylabel('<Etot> [J]')
plt.title('Energia totale media')
plt.grid(True)
plt.show()
```
