# Laboratorio di simulazioni numeriche 
### Esercizio 10.1

Si richiede di adattare l'algoritmo genetico (GA) di **Esercizio 9.1** per risolvere il problema del commesso viaggiatore tramite simulated annealing (SA) nelle configurazioni:
- 32 città disposte in modo random su una circonferenza di raggio unitario;
- 32 città disposte in modo random all'interno di un quadrato di lato unitario.

Di seguito si riportano i percorsi ottimali ottenuti nelle due configurazioni con il codice SA e la lunghezza del percorso ottimale in funzione delle generazioni. I risultati sono confrontati con quelli ottenuti con il codice GA.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

data_SA=np.loadtxt('Lab10/data_circ.dat')
L_circ_SA=np.loadtxt('Lab10/L_circ.dat')
x1=data_SA[:,0]
y1=data_SA[:,1]
num=data_SA[:,2]
n1=np.arange(0,len(L_circ_SA))

data_GA=np.loadtxt('Lab9/data_circ.dat')
L_circ_GA=np.loadtxt('Lab9/L_circ.dat')
x2=data_GA[:,0]
y2=data_GA[:,1]
num=data_GA[:,2]
n2=np.arange(0,len(L_circ_GA))

plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.axis('scaled')
plt.title('Percorso ottimale SA')

plt.subplot(2,2,2)
plt.plot(n1,L_circ_SA[:])
plt.title('Costo ottimale SA')
plt.xlabel('generazione')

plt.subplot(2,2,3)
plt.plot(x2,y2)
plt.axis('scaled')
plt.title('Percorso ottimale GA')

plt.subplot(2,2,4)
plt.plot(n2,L_circ_GA[:,0])
plt.title('Costo ottimale GA')
plt.xlabel('generazione')

plt.show()
```

```python
data_SA=np.loadtxt('Lab10/data_square.dat')
L_square_SA=np.loadtxt('Lab10/L_square.dat')
x1=data_SA[:,0]
y1=data_SA[:,1]
num=data_SA[:,2]
n1=np.arange(0,len(L_square_SA))

data_GA=np.loadtxt('Lab9/data_square.dat')
L_square_GA=np.loadtxt('Lab9/L_square.dat')
x2=data_GA[:,0]
y2=data_GA[:,1]
num=data_GA[:,2]
n2=np.arange(0,len(L_square_GA))

plt.figure(figsize=(16,10))
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.axis('scaled')
plt.title('Percorso ottimale SA')

plt.subplot(2,2,2)
plt.plot(n1,L_square_SA[:])
plt.title('Costo ottimale SA')
plt.xlabel('generazione')

plt.subplot(2,2,3)
plt.plot(x2,y2)
plt.axis('scaled')
plt.title('Percorso ottimale GA')

plt.subplot(2,2,4)
plt.plot(n2,L_square_GA[:,0])
plt.title('Costo ottimale GA')
plt.xlabel('generazione')

plt.show()
```

I risultati mostrano come a parità di configurazioni il SA permetta di trovare il percorso ottimale più facilmente del semplice GA (particolamente evidente dalla figura del cerchio). Il SA infatti , a differenza del GA, permette di indirizzare la ricerca verso il minimo globale, limitando il rischio di rimanere intrappolati in minimi locali. Osserviamo inoltre che il costo ottimale sembra più stabile nel caso del SA, specie nella fase finale delle simulazioni: ciò potrebbe segnalare che raggiunte configurazioni ottimali, l'abbassarsi della "temperatura" blocchi il rischio di uscire dal minimo cercato tramite mutazioni casuali.


### Esercizio 10.2

L'esercizio prevede di parallelizzare tramite le librerie MPI il GA di **Esercizio 9.1**. Ognuno dei 4 nodi deve operare una ricerca indipendente (tramite diversa inizializzazione del generatore di numeri random) ma ogni $N_{migr}$ vengono selezionate coppie di due nodi (continenti) per scambiare il miglior individuo delle rispettive generazioni.

Osservazione: piuttosto che una selezione random dei continenti per effettuare lo scambio si è preferito operare tramite una selezione deterministica per valutare se questa scelta possa migliorare la convergenza dell'algoritmo. Per questo motivo sono stati alternativamente scambiati gli individui tra $0-1$, $2-3$ e tra $0-3$, $1-2$. 

In ogni caso, per effettuare la selezione random sarebbe bastato effettuare una BCast dal nodo 0 agli altri per avere una selezione concorde dei continenti scelti.

Vengono di seguito riportati il miglior cammino per ogni continente e l'andamento della lunghezza minima di ogni generazione, sia per le città posizionate sulla circonferenza che per quelle dentro il quadrato.

```python
data0=np.loadtxt('Lab10par/data0.dat')
data1=np.loadtxt('Lab10par/data1.dat')
data2=np.loadtxt('Lab10par/data2.dat')
data3=np.loadtxt('Lab10par/data3.dat')
L0=np.loadtxt('Lab10par/dataL0.dat')
L1=np.loadtxt('Lab10par/dataL1.dat')
L2=np.loadtxt('Lab10par/dataL2.dat')
L3=np.loadtxt('Lab10par/dataL3.dat')

plt.figure(figsize=(16,21))
plt.subplot(4,2,1)
plt.plot(data0[:,0],data0[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 0')

n0=np.arange(0,len(L0[:,0]))
plt.subplot(4,2,2)
plt.plot(n0,L0[:,0])
plt.title('Costo ottimale continente 0')
plt.xlabel('generazione')

plt.subplot(4,2,3)
plt.plot(data1[:,0],data1[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 1')

n1=np.arange(0,len(L1[:,0]))
plt.subplot(4,2,4)
plt.plot(n1,L1[:,0])
plt.title('Costo ottimale continente 1')
plt.xlabel('generazione')

plt.subplot(4,2,5)
plt.plot(data2[:,0],data2[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 2')

n2=np.arange(0,len(L2[:,0]))
plt.subplot(4,2,6)
plt.plot(n2,L2[:,0])
plt.title('Costo ottimale continente 2')
plt.xlabel('generazione')

plt.subplot(4,2,7)
plt.plot(data3[:,0],data3[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 3')

n3=np.arange(0,len(L3[:,0]))
plt.subplot(4,2,8)
plt.plot(n3,L3[:,0])
plt.title('Costo ottimale continente 3')
plt.xlabel('generazione')

plt.show()
```

```python
data0=np.loadtxt('Lab10par/data0_s.dat')
data1=np.loadtxt('Lab10par/data1_s.dat')
data2=np.loadtxt('Lab10par/data2_s.dat')
data3=np.loadtxt('Lab10par/data3_s.dat')
L0=np.loadtxt('Lab10par/dataL0_s.dat')
L1=np.loadtxt('Lab10par/dataL1_s.dat')
L2=np.loadtxt('Lab10par/dataL2_s.dat')
L3=np.loadtxt('Lab10par/dataL3_s.dat')

plt.figure(figsize=(16,21))
plt.subplot(4,2,1)
plt.plot(data0[:,0],data0[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 0')

n0=np.arange(0,len(L0[:,0]))
plt.subplot(4,2,2)
plt.plot(n0,L0[:,0])
plt.title('Costo ottimale continente 0')
plt.xlabel('generazione')

plt.subplot(4,2,3)
plt.plot(data1[:,0],data1[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 1')

n1=np.arange(0,len(L1[:,0]))
plt.subplot(4,2,4)
plt.plot(n1,L1[:,0])
plt.title('Costo ottimale continente 1')
plt.xlabel('generazione')

plt.subplot(4,2,5)
plt.plot(data2[:,0],data2[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 2')

n2=np.arange(0,len(L2[:,0]))
plt.subplot(4,2,6)
plt.plot(n2,L2[:,0])
plt.title('Costo ottimale continente 2')
plt.xlabel('generazione')

plt.subplot(4,2,7)
plt.plot(data3[:,0],data3[:,1])
plt.axis('scaled')
plt.title('Percorso ottimale continente 3')

n3=np.arange(0,len(L3[:,0]))
plt.subplot(4,2,8)
plt.plot(n3,L3[:,0])
plt.title('Costo ottimale continente 3')
plt.xlabel('generazione')

plt.show()
```

I risultati mostrano come pur potendo comunicare tra loro, i percorsi ottimali dei 4 continenti non sono necessariamente gli stessi. 

La possibilità di operare su più nodi migliora sicuramente il raggiungimento del percorso ottimale: nel caso del cerchio infatti è evidente come questo sia stato trovato in 2 casi su 4. Osservando lo spazio lasciato tra la prima e l'ultima città si può notare come questi 2 percorsi siano stati trovati uno in senso orario e uno in senso antiorario: la parallelizzazione ha quindi permesso di esplorare e scambiare percorsi ottimali tramite due modi di convergenza diversi.

La scelta di effettuare scambi deterministici inoltre ha prodotto un maggiore grado di somiglianza (e quindi di convergenza) tra i continenti comunicanti direttamente: è evidente infatti come il percorso $0$ sia particolarmente simile a $1$ e $4$ in entrambi i casi, ma significativamente diverso rispetto a $2$. Il metodo scelto non permette quindi di migliorare la convergenza a livello globale ma di affinarne il grado su singoli blocchi di percorso.
