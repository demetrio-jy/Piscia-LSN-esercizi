### Esercizio 8.1 e 8.2
Il ground state di una particella vincolata dal potenziale in 1D

$$
V(x) = x^4 - \frac{5}{2}x^2
$$

può essere stimato attraverso un metodo variazionale Monte Carlo assumendo come funzione d'onda di prova 

$$
\Psi_T(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

Per fare ciò è necessario campionare la distribuzione di probabilità $|\Psi_T(x)|^2$ con l'algoritmo di Metropolis (si assume probabilità di transizione uniforme), in modo che il valor medio dell'energia sia stimabile dall'alto attraverso 

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

Variando i parametri della funzione d'onda di prova è stato trovato che l'energia media risulta minimizzata per $\mu = 0.8$ e $\sigma^2 = 0.38$. Vengono riportate di seguito la stima di $\langle {\hat H} \rangle_T$ ricavata attraverso il metodo della media a blocchi e la distribuzione campionata. Quest'ultima viene confrontata con la distribuzione di prova fornita $|\Psi_T(x)|^2$ e con quella ricavata attraverso soluzione numerica dell'equazione di Schrodinger $|\Psi(x)|^2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # step size
V = Vpot(x)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1) # central differences method
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V) # hamiltonian matrix
E,psi = np.linalg.eigh(H) # eigenvectors and eigenvalues
psi = np.transpose(psi) # transpose 
psi = psi/np.sqrt(dx) # normalize

print("L'energia del ground state calcolata numericamente è: ", E[0])
print("L'energia del primo stato eccitato è: ", E[1])
print("l'energia del secondo stato eccitato è: ", E[2])

mu = 0.8
s2 = 0.38

def psi_T2(x):
    return (np.exp(-(x-mu)**2/(2*s2)) + np.exp(-(x+mu)**2/(2*s2)))**2

norm = integrate.quad(lambda x: psi_T2(x), -4, 4)
psi_T = np.exp(-(x-mu)**2/(2*s2)) + np.exp(-(x+mu)**2/(2*s2))
```

```python
energy = np.loadtxt("Lab8a/mean.dat")
mean = energy[:,0]
error = energy[:,1]
n = np.arange(100) 

conf = np.loadtxt("Lab8a/configuration.dat")

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
e_graph, psi_graph = axes.flatten()

e_graph.errorbar(n,mean,yerr=error)
e_graph.axhline(y=-0.46046588,color='red',label='Energia GS teorica')
e_graph.set_title('energia media variazionale')
e_graph.set(xlabel='#block', ylabel='$<H>_T$')
e_graph.legend()
e_graph.grid(True)

psi_graph.hist(conf, bins=100, density=True)
psi_graph.set_title('campionamento & distribuzioni')
l1 = psi_graph.plot(x,(psi[0])**2, color='orange')
l2 = psi_graph.plot(x,(psi_T)**2/norm[0], color='red')
psi_graph.set_xlim((-3,3))
psi_graph.set(xlabel="x")
psi_graph.grid(True)

line_labels = ["$|\psi|^2$", "$|\psi_T|^2$"]
psi_graph.legend([l1, l2], labels=line_labels)

plt.show()
```

I risulttati mostrano come la distribuzione campionata riproduca la distribuzione di prova $|\Psi_T(x)|^2$ ma si discosti significativamente da quella teorica $|\Psi(x)|^2$. L'energia media ottenuta inoltre risulta maggiore rispetto a quella stimata con soluzione numerica. Ciò è coerente con il fatto che il metodo seguito è di tipo variazionale.


### Esercizio 8.3

Utilizzando il codice PIGS/PIMC fornito, viene richiesto di studiare il sistema unidimensionale, inizialmente a temperatura nulla, utilizzando delle funzioni d'onda di prova:
- uniforme $\Psi_T(x)=1$
- doppia gaussiana $\Psi_T(x) = \Psi_T^{\sigma,\mu}(x)$ con i parametri ottimizzati $\mu=0.8$, $\sigma^2=0.38$

I risultati sono confrontati con quelli del codice VMC degli esercizi precedenti.

```python
ep_uni = np.loadtxt("Lab8b/potential.uniform.dat")
ek_uni = np.loadtxt("Lab8b/kinetic.uniform.dat")
prob_uni = np.loadtxt("Lab8b/probability.uniform.dat")

ep_pigs = np.loadtxt("Lab8b/potential.PIGS.dat")
ek_pigs = np.loadtxt("Lab8b/kinetic.PIGS.dat")
prob_pigs = np.loadtxt("Lab8b/probability.PIGS.dat")

etot_uni = ep_uni[:,1] + ek_uni[:,1]
etot_uni_err = np.sqrt(ep_uni[:,2]**2 + ek_uni[:,2]**2)

etot_pigs = ep_pigs[:,1] + ek_pigs[:,1]
etot_pigs_err = np.sqrt(ep_pigs[:,2]**2 + ek_pigs[:,2]**2)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 11))
eu_graph, pu_graph , ep_graph, pp_graph = axes.flatten()

eu_graph.errorbar(ep_uni[:,0],etot_uni,yerr=etot_uni_err)
eu_graph.axhline(y=-0.46046588,color='red',label='Energia GS teorica')
eu_graph.set_title('energia media, funzione uniforme')
eu_graph.set(xlabel='#timeslice', ylabel='$<H>_T$')
eu_graph.legend()
eu_graph.grid(True)

pu_graph.set_title('distribuzione con funzione uniforme')
l1 = pu_graph.plot(x,(psi[0])**2, color='blue')
l2 = pu_graph.plot(x,(psi_T)**2/norm[0], color='orange')
l3 = pu_graph.plot(prob_uni[:,0],prob_uni[:,1], color='red')
pu_graph.set_xlim((-3,3))
pu_graph.set(xlabel="x")
pu_graph.grid(True)
line_labels = ["$|\psi|^2$", "$|\psi_{vcm}|^2$", "$|\psi_{pigs}|^2$"]
pu_graph.legend([l1, l2, l3], labels=line_labels)

ep_graph.errorbar(ep_pigs[:,0],etot_pigs,yerr=etot_pigs_err)
ep_graph.axhline(y=-0.46046588,color='red',label='Energia GS teorica')
ep_graph.set_title('energia media, funzione data')
ep_graph.set(xlabel='#timeslice', ylabel='$<H>_T$')
ep_graph.legend()
ep_graph.grid(True)

pp_graph.set_title('distribuzione con funzione data')
l1 = pp_graph.plot(x,(psi[0])**2, color='blue')
l2 = pp_graph.plot(x,(psi_T)**2/norm[0], color='orange')
l3 = pp_graph.plot(prob_pigs[:,0],prob_pigs[:,1], color='red')
pp_graph.set_xlim((-3,3))
pp_graph.set(xlabel="x")
pp_graph.grid(True)
line_labels = ["$|\psi|^2$", "$|\psi_{vcm}|^2$", "$|\psi_{pigs}|^2$"]
pp_graph.legend([l1, l2, l3], labels=line_labels)

plt.show()
```

Entrambe le funzioni di prova danno luogo a distribuzioni di probabilità e valori medi dell'energia totale analoghi. Sia per la funzione di prova uniforme sia per $\Psi_T^{\sigma,\mu}(x)$ la distribuzione ottenuta approssima ottimamente la distribuzione ottenuta con risoluzione analitica. 
Entrambe le simulazioni riportate si riferiscono ad un tempo immaginario di propagazione pari a 8.
Si osserva che abbassando il tempo immaginario sotto 2 gli errori nelle energie aumentano. 

Successivamente viene richiesto di confrontare i risultati ottenuti a temperatura nulla con quelli a temperatura finita. Sono state prese in considerazioni le seguenti configurazioni:

#### Temperatura   $T=1.25$

```python
ep_pimc = np.loadtxt("Lab8b/potential.PIMC.dat")
ek_pimc = np.loadtxt("Lab8b/kinetic.PIMC.dat")
prob_pimc = np.loadtxt("Lab8b/probability.PIMC.dat")

etot_pimc = ep_pimc[:,1] + ek_pimc[:,1]
etot_pimc_err = np.sqrt(ep_pimc[:,2]**2 + ek_pimc[:,2]**2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
e_graph, p_graph = axes.flatten()

e_graph.errorbar(ep_pimc[:,0],etot_pimc,yerr=etot_pimc_err)
e_graph.axhline(y=0.097424713,color='red',label='Energia 1 ecc. teorica')
e_graph.set_title('energia media, funzione data')
e_graph.set(xlabel='#timeslice', ylabel='$<H>_T$')
e_graph.legend(loc=0)
e_graph.grid(True)

p_graph.set_title('distribuzione con funzione data')
l1 = p_graph.plot(prob_pigs[:,0],prob_pigs[:,1], color='orange')
l2 = p_graph.plot(prob_pimc[:,0],prob_pimc[:,1], color='red')
p_graph.set_xlim((-3,3))
p_graph.set(xlabel="x")
p_graph.grid(True)
line_labels = ["$|\psi_{pigs}|^2$", "$|\psi_{pimc}|^2$"]
p_graph.legend([l1, l2], labels=line_labels)

plt.show()
```

#### Temperatura $T=3$

```python
ep_pimc = np.loadtxt("Lab8b/potential.PIMC3.dat")
ek_pimc = np.loadtxt("Lab8b/kinetic.PIMC3.dat")
prob_pimc = np.loadtxt("Lab8b/probability.PIMC3.dat")

etot_pimc = ep_pimc[:,1] + ek_pimc[:,1]
etot_pimc_err = np.sqrt(ep_pimc[:,2]**2 + ek_pimc[:,2]**2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
e_graph, p_graph = axes.flatten()

e_graph.errorbar(ep_pimc[:,0],etot_pimc,yerr=etot_pimc_err)
e_graph.axhline(y=1.971733134,color='orange',label='Energia 2 ecc. teorica')
e_graph.axhline(y=0.097424713,color='red',label='Energia 1 ecc. teorica')
e_graph.set_title('energia media, funzione data')
e_graph.set(xlabel='#timeslice', ylabel='$<H>_T$')
e_graph.legend(loc=0)
e_graph.grid(True)

p_graph.set_title('distribuzione con funzione data')
l1 = p_graph.plot(prob_pigs[:,0],prob_pigs[:,1], color='orange')
l2 = p_graph.plot(prob_pimc[:,0],prob_pimc[:,1], color='red')
p_graph.set_xlim((-3,3))
p_graph.set(xlabel="x")
p_graph.grid(True)
line_labels = ["$|\psi_{pigs}|^2$", "$|\psi_{pimc}|^2$"]
p_graph.legend([l1, l2], labels=line_labels)

plt.show()
```

I risultati mostrano come a temperatura finita si registri un'espansione delle distribuzioni verso le code e un aumento dell'energia media del sistema fino a raggiungere valori positivi. All'aumentare della temperatura tali fenomeni si intensificano. Infatti l'effetto della temperatura è quello di mischiare il ground state con gli stati eccitati, con il peso dei secondi che aumenta con la temperatura.

In effetti alla temperatura $T=1.25$ il comportamento a $x=0$ può essere indice del fatto che lo stato maggiormente eccitato è il primo (che ha un nodo nell'origine), mentre a $T=3$ l'altezza del minimo aumenta come conseguenza del fatto che aumenta il peso del secondo stato eccitato (che nell'origine ha un massimo).
