# Laboratorio di simulazioni numeriche
### Esercizio 2.1
L'esercizio richiede di calcolare l'integrale 

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

utilizzando un campionamento di tipo Monte Carlo. Per fare ciò l'integrale viene assunto della forma 

$$I = \int_0^1 g(x) p(x) dx = 1$$

e si richiede di effettuare un campionamento di $x$ nell'intervallo $[0,1]$ di tipo:

- uniforme: $p(x)=1$ e $g(x)=\pi/2\cos(\pi x/2)$;
- tramite important sampling.

Nel secondo caso la distribuzione scelta è stata $p(x)=2(1-x)$, in modo tale che $g(x)=\frac{\pi}{4(1-x)}\cos(\pi x/2)$.

Vengono di seguito riportati i valori medi ottenuti in funzione del numero di blocchi utilizzati nel blocking method. Il numero di campionamenti usati è in entrambi i casi $M=10^5$, la dimensioni dei blocchi è di $1000$ campionamenti ciascuno.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

uniform = np.loadtxt("Lab2/integral.dat")
nblock_u = uniform[:,0]/1000.
mean_u = uniform[:,1]
error_u = uniform[:,2]

importance = np.loadtxt("Lab2/importance.dat")
nblock_i = importance[:,0]/1000.
mean_i = importance[:,1]
error_i = importance[:,2]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
uni, imp = axes.flatten()

uni.errorbar(nblock_u,mean_u,yerr=error_u)
uni.axhline(y=1,color='red')
uni.set_title('Campionamento uniforme')
uni.set(xlabel='#block')
uni.grid(True)

imp.errorbar(nblock_i,mean_i,yerr=error_i)
imp.axhline(y=1,color='red')
imp.set_title('Importance sampling')
imp.set(xlabel='#block')
imp.grid(True)

plt.show()
```

I risultati mostrano come a parità di configurazione, l'importance sampling risulti un metodo più accurato per calcolare l'integrale.

### Esercizio 2.2

Viene richiesto di implementare un codice per il random walk 3D 
- nel caso di moto su reticolo cubico di passo $a=1$ (step di lunghezza 1 in ognuna delle direzioni spaziali);
- nel caso continuo (step di lunghezza 1 in direzione random $\theta \in [0,\pi]$ e $\phi \in [0,2\pi]$, con angoli campionati uniformemente).

In entrambi i casi vengono svolte $M=10^4$ simulazioni di $N=100$ step ciascuna, ogni volta con partenza nell'origine.

Vengono riportate di seguito plot della media quadratica (RMS) della posizione $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ in funzione degli step per entrambi i random walk.

```python
lattice = np.loadtxt("Lab2/RW1.dat")
mean_l = lattice[:,0]
error_l = lattice[:,1]
mean_l = np.sqrt(mean_l)
error_l = error_l/(2*mean_l)

continuous = np.loadtxt("Lab2/RW2.dat")
mean_c = continuous[:,0]
error_c = continuous[:,1]
mean_c = np.sqrt(mean_c)
error_c = error_c/(2*mean_c)

n = np.arange(100) +1

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))
lat, cont = axes.flatten()

lat.plot(n,np.sqrt(n),color='r',linewidth = 3)
lat.errorbar(n,mean_l,yerr=error_l)
lat.set_title('RW su reticolo')
lat.set(xlabel='#step', ylabel='RMS')
lat.grid(True)

cont.plot(n,np.sqrt(n),color='r',linewidth = 3)
cont.errorbar(n,mean_c,yerr=error_c)
cont.set_title('RW continuo')
cont.set(xlabel='#step', ylabel='RMS')
cont.grid(True)

plt.show()
```

L'andamento qualitativo dei due random walk è lo stesso: in entrambi i casi la RMS riproduce con ottima approssimazione quello della funzione $f(N)=\sqrt{N}$ (in rosso nei grafici).

Tale comportamento è indice di un moto di tipo diffusivo, come ci si aspetta da un random walk.
