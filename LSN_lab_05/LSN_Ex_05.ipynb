### Orbitale 1s

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

R1s_t = 1.5
n = np.arange(1000)

file1 = np.loadtxt("Lab5/data1.dat")
R1s_u = file1[:,0]
Err_R1s_u = file1[:,1]

file2 = np.loadtxt("Lab5/data3.dat")
R1s_g = file2[:,0]
Err_R1s_g = file2[:,1]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
u_graph, g_graph = axes.flatten()

u_graph.errorbar(n,R1s_u-R1s_t,yerr=Err_R1s_u)
u_graph.set_title('transizione uniforme')
u_graph.set(xlabel='#block', ylabel='<r> - <r>_teo')
u_graph.grid(True)

g_graph.errorbar(n,R1s_g-R1s_t,yerr=Err_R1s_g)
g_graph.set_title('transizione gaussiana')
g_graph.set(xlabel='#block', ylabel='<r> - <r>_teo')
g_graph.grid(True)

plt.show()
```

### Orbitale 2p

```python
R1s_t = 5

file1 = np.loadtxt("Lab5/data2.dat")
R2p_u = file1[:,0]
Err_R2p_u = file1[:,1]

file2 = np.loadtxt("Lab5/data4.dat")
R2p_g = file2[:,0]
Err_R2p_g = file2[:,1]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
u_graph, g_graph = axes.flatten()

u_graph.errorbar(n,R2p_u-R1s_t,yerr=Err_R2p_u)
u_graph.set_title('transizione uniforme')
u_graph.set(xlabel='#block', ylabel='<r> - <r>_teo')
u_graph.grid(True)

g_graph.errorbar(n,R2p_g-R1s_t,yerr=Err_R2p_g)
g_graph.set_title('transizione gaussiana')
g_graph.set(xlabel='#block', ylabel='<r> - <r>_teo')
g_graph.grid(True)

plt.show()
```

Le simulazioni sono state svolte per un numero di passi $M=10^6$ e le medie dei raggi sono state effettuate su blocchi da $100$ elementi ciascuno. Per avere un rate di accettazione del $50\%$ sono stati utilizzati: 
- $\delta = 1.223$ e $\sigma = 0.764$ per $\Psi_{1,0,0}$;
- $\delta = 3.18$ e $\sigma = 1.996$ per $\Psi_{2,1,0}$;

In un primo momento le simulazioni sono state effettuate partendo da una posizione molto lontana dall'origine, $\vec y = (100;100;100)$, ed è stato osservato come l'algoritmo di Metropolis favorisca nel giro di $200$ blocchi circa l'esplorazione di una regione di spazio vicina all'origine, per poi stabilizzarsi. La partenza da un punto molto lontano ha comunque l'effetto di disturbare molto il valore del valor medio e l'incertezza dei raggi. 

Per questo motivo si è scelto di partire da una posizione $\vec y = (10;10;10)$, in modo da poter osservare il transiente iniziale senza compromettere troppo i dati.

Osserviamo che la scelta di operare con blocchi di $100$ step cascuno è stata motivata più dal voler evidenziare il transiente iniziale che dal voler ottenere una precisione maggiore. Utilizzando blocchi da $1000$ step infatti è possibile osservare fluttuazioni più contenute, confrontabili con la seconda cifra decimale. D'altra parte l'utilizzo di blocchi troppo grandi è sconsigliabile in quanto aumenta la correlazione tra vari blocchi.

Per quanto riguarda l'utilizzo delle diverse distribuzioni per effettuare le transizioni, non si osservano variazioni significative nel passaggio da uniforme a gaussiana. In effetti più che il tipo di distribuzione, il fattore determinante nel campionamento dei dati è il rate di accettazione: avendo un rate di circa il $50\%$ in entrambi i casi, mi aspetto che le due distribuzioni producano campionamenti confrontabili.
