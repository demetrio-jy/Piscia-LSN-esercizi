# Laboratorio di simulazioni numeriche
### Esercizio 6.1

In questa esercitazione viene richiesto di campionare un sistema monodimensionale di spin $\{ s_i \}_{i=1,N}$, descritti dal modello di Ising a primi vicini tramite l'hamiltoniana

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-h\sum_{i=1}^N s_i 
$$

utilizzando alternativamente l'algoritmo di Metropolis e quello di Gibbs. La statistica del sistema viene espressa in termini della funzione di partizione 

$$
Z=\sum_{\{ s_i \}} \exp \left( - \beta H  \right)
$$

dove $\beta = 1/ k_b T$, $T$ è la temperatura del sistema e $k_b = 1$. Il campionamento fa riferimento pertanto alla distribuzione di probabilità di Boltzmann 

$$
p(\{ s_i \}) = \exp(-\beta H)/Z
$$

Di seguito vengono confrontati il valore teorico delle grandezze fisiche del sistema e i valori medi delle simulazioni, ottenuti tramite il metodo della media a blocchi, in funzione della tempertaura $T$. Laddove non specificato, le simulazioni sono state svolte a campo esterno $h$ nullo. Il numero di spin è stato fissato a $N = 50$ con condizioni al bordo periodiche e accoppiamento $J = 1$.

### Energia interna 

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
import numpy as np
import matplotlib.pyplot as plt

metro1 = np.loadtxt("Lab6/data_metro.dat")
T_data = metro1[:,0]
U_m = metro1[:,1]
err_U_m = metro1[:,2]
c_m = metro1[:,3]
err_c_m = metro1[:,4]
x_m = metro1[:,5]
err_x_m = metro1[:,6]

metro2 = np.loadtxt("Lab6/mag_metro.dat")
M_m = metro2[:,1]
err_M_m = metro2[:,2]

gibbs1 = np.loadtxt("Lab6/data_gibbs.dat")
U_g = gibbs1[:,1]
err_U_g = gibbs1[:,2]
c_g = gibbs1[:,3]
err_c_g = gibbs1[:,4]
x_g = gibbs1[:,5]
err_x_g = gibbs1[:,6]

gibbs2 = np.loadtxt("Lab6/mag_gibbs.dat")
M_g = gibbs2[:,1]
err_M_g = gibbs2[:,2]
```

```python
points=100
T = np.linspace(0.45,2.05,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
m_graph, g_graph = axes.flatten()

m_graph.plot(T, e,label="Valore teorico")
m_graph.errorbar(T_data,U_m,yerr=err_U_m,fmt='o',label="Campionamento")
m_graph.set_title('Metropolis sampling')
m_graph.set(xlabel='T', ylabel='U/N')
m_graph.legend(loc=0)

g_graph.plot(T, e,label="Valore teorico")
g_graph.errorbar(T_data,U_g,yerr=err_U_g,fmt='o',label="Campionamento")
g_graph.set_title('Gibbs sampling')
g_graph.set(xlabel='T', ylabel='U/N')
g_graph.legend(loc=0)

plt.show()
```

### Calore specifico

$$
\frac{C(N,T)}{N} = 
\frac{k_B\beta^2}{N} \left(\langle H^2 \rangle -\langle H \rangle^2 \right) = \\ \underset{h=0} =
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
m_graph, g_graph = axes.flatten()

m_graph.plot(T, heat,label='Valore teorico')
m_graph.errorbar(T_data,c_m,yerr=err_c_m,fmt='o',label='Campionamento')
m_graph.set_title('Metropolis sampling')
m_graph.set(xlabel='T', ylabel='c/N')
m_graph.legend(loc=0)

g_graph.plot(T, heat,label='Valore teorico')
g_graph.errorbar(T_data,c_g,yerr=err_c_g,fmt='o',label='Campionamento')
g_graph.set_title('Gibbs sampling')
g_graph.set(xlabel='T', ylabel='c/N')
g_graph.legend(loc=0)

plt.show()
```

### Magnetizzazione (h = 0.02)

$$
\frac{M(N,T,h)}{N} = \frac{1}{N} \left\langle \sum_{i=1,N}s_i \right\rangle = \\ = \frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

$$
\lambda_{1,2} = e^{\beta J} \cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}
$$

```python
h=0.02
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
m_graph, g_graph = axes.flatten()

m_graph.plot(T, M,label='Valore teorico')
m_graph.errorbar(T_data,M_m,yerr=err_M_m,fmt='o',label='Campionamento')
m_graph.set_title('Metropolis sampling')
m_graph.set(xlabel='T', ylabel='M/N')
m_graph.legend(loc=0)

g_graph.plot(T, M,label='Valore teorico')
g_graph.errorbar(T_data,M_g,yerr=err_M_g,fmt='o',label='Campionamento')
g_graph.set_title('Gibbs sampling')
g_graph.set(xlabel='T', ylabel='M/N')
g_graph.legend(loc=0)

plt.show()
```

### Suscettività magnetica

$$
\chi(N,T) \underset{h=0} = 
 \frac{\beta}{N} \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right]  = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
m_graph, g_graph = axes.flatten()

m_graph.plot(T, X,label='Valore teorico')
m_graph.errorbar(T_data,x_m,yerr=err_x_m,fmt='o',label='Campionamento')
m_graph.set_title('Metropolis sampling')
m_graph.set(xlabel='T', ylabel='$\chi$')
m_graph.legend(loc=0)

g_graph.plot(T, X,label='Valore teorico')
g_graph.errorbar(T_data,x_g,yerr=err_x_g,fmt='o',label='Campionamento')
g_graph.set_title('Gibbs sampling')
g_graph.set(xlabel='T', ylabel='$\chi$')
g_graph.legend(loc=0)

plt.show()
```

I risultati mostrano buon accordo tra simulazioni e previsioni teoriche per la maggior parte dei casi in esame, specialmente a temperature più elevate (dove l'inversione degli spin dovuta al campionamento non ha un effetto significativo sulle proprietà del sistema). Osserviamo che i due diversi tipi di campionamento producono risultati confrontabili.
