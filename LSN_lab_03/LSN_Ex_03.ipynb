# Laboratorio di simulazioni numeriche
### Esercizio 3.1
Viene richiesto di stimare il prezzo di opzioni Call e Put di tipo europeo a un tempo $T$ attraverso l'utilizzo di: 

1) sampling diretto dei valori $C[S(T),T]$, $P[S(T),T]$ ottenibili attraverso la teoria del modello di Black-Scholes:

$$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$

   dove $W(T)\sim\mathcal{N}(0,T)$;

2) campionamento dei prezzi in intervalli successivi $t_i$, con $t_0 = 0$  e  $t_{100} = T$, sapendo che  

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
   dove $Z_i \sim \mathcal{N}(0,1)$.

Le simulazioni vengono effettuate utilizzando i seguenti parametri:
- prezzo a $t=0$: $S(0)=100$
- tempo finale: $T=1$
- prezzo d'esercizio: $K=100$
- interesse risk-free: $r=0.1$
- volatilit√†: $\sigma=0.25$

I valori dei prezzi vengono confrontati con quelli ottenuti tramite calcolo analitico:

Call:  14.975790778311286

Put:  5.4595325819072364

### Sampling diretto

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

C_th = 14.975790778311286
P_th = 5.4595325819072364

file = np.loadtxt("Lab3/data1.dat")
n = file[:,0]
C = file[:,1]
C_error = file[:,2]
P = file[:,3]
P_error = file[:,4]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
C_graph, P_graph = axes.flatten()

C_graph.errorbar(n,C-C_th,yerr=C_error)
C_graph.set_title('opzione Call')
C_graph.set(xlabel='#block', ylabel='C[S(T),T] - C_th')
C_graph.grid(True)

P_graph.errorbar(n,P-P_th,yerr=P_error)
P_graph.set_title('opzione Put')
P_graph.set(xlabel='#block', ylabel='P[S(T),T] - P_th')
P_graph.grid(True)

plt.show()
```

### Campionamento a intervalli successivi

```python
file = np.loadtxt("Lab3/data2.dat")
n = file[:,0]
C = file[:,1]
C_error = file[:,2]
P = file[:,3]
P_error = file[:,4]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
C_graph, P_graph = axes.flatten()

C_graph.errorbar(n,C-C_th,yerr=C_error)
C_graph.set_title('opzione Call')
C_graph.set(xlabel='#block', ylabel='C[S(T),T] - C_th')
C_graph.grid(True)

P_graph.errorbar(n,P-P_th,yerr=P_error)
P_graph.set_title('opzione Put')
P_graph.set(xlabel='#block', ylabel='P[S(T),T] - P_th')
P_graph.grid(True)

plt.show()
```
